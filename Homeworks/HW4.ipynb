```python
class word():
    def __init__(self, size, letters):
        self.size = size
        self.letters = letters
        
letters = input(str("Player 1, please enter your word: "))
size = len(letters)

askedWord = word(size, letters)

displayedWord = "_"*askedWord.size

livesRemain = 6

print("Player 2, the word that Player 1 asks has", askedWord.size, "letters.")

theWord = list(askedWord.letters)
theDisplay = list(displayedWord)
print(theDisplay,"\n")
print("IMPORTANT NOTE FOR PLAYER 2: If your guess is in the word for multiple times, you are able to see only one of it.")
while livesRemain > 0:
        guess = input(str("Player 2, can you guess the letters? Enter a letter: "))
        if guess in theWord:
            print("Yes! You are correct! The letter  '", guess,"'  is in the word.")
            theDisplay[theWord.index(guess)] = guess
            print(theDisplay,"\n")
           
        else:
            print("No! The letter  '", guess,"'  is not in the word!")
            livesRemain -= 1
            print("You have ",livesRemain,"attempt(s) left.")
        
        if theDisplay == theWord:
            print("Congratulations! Player 2 won the game!")
            break
        if livesRemain == 0:
            print("Player 1 wins!")
            break
```
